## **Iteration 3: Temperature= 0.05 and Max_tokens = 2000 + output format description**

```python
prompt = f"""
Given the following dataset:
{dataset}

You are a forecasting software.
Take the date from the index of the dataset. Calculate a daily forecast for 'Price' for the period **2021-01-01 to 2021-03-31** using an ARMA model and using the history in {dataset} from **2020-01-03 to 2020-12-30** to fit the model.

1. Handle any missing information or NaNs in the dataset.
2. Return **only the results** in a complete table with the columns: 'Date', 'Price', 'Forecasted_Price', and 'Residuals'.
   - The table must include **all dates** from **2021-01-01 to 2021-03-31**.
   - Ensure that the 'Residuals' column is calculated as: Residuals = Price - Forecasted_Price.
   - Use this exact format, but don't use the values that appear on it:
      Date,Price,Forecasted_Price,Residuals
3. After printing the table, calculate and display the values of **RMSE**, **MAE**, and **AIC** for the ARMA model.
4. Your output must **not include any Python code or explanations**. It must only return the requested table and metrics.
Ensure that the table includes **all dates** from **2021-01-01 to 2021-03-31**, even if some dates are missing in the original dataset. Fill in any missing dates with the appropriate forecasts and residuals.
"""
```

```python
import requests
import json

# Configure the request to the Ollama API
url = "http://localhost:11434/api/generate"
payload = {
    "model": "deepseek-r1:1.5b",  # Name of the model you downloaded
    "prompt": prompt,
    "temperature": 0.05,  # Adjust the temperature (0.0 to 1.0)
    "max_tokens": 2000,  # Set the maximum number of tokens in the response
    "stream": False  # If you want a real-time response, change to True
}

# Send the request
response = requests.post(url, json=payload)

# Process the response
if response.status_code == 200:
    result = response.json()
    print("Model answer:")
    print(result["response"])
else:
    print(f"Error: {response.status_code} - {response.text}")
```
